```python
import pandas as pd

# Read the first CSV file
telecom_df = pd.read_csv("telecom_dataset.csv")

# Read the second CSV file
energy_df = pd.read_csv("Energy_dataset.csv")

# Concatenate the DataFrames
combined_df = pd.concat([telecom_df, energy_df])

# Check the number of rows in individual DataFrames
print("Number of rows in telecom_df:", len(telecom_df))
print("Number of rows in energy_df:", len(energy_df))

# Check the number of rows in the combined DataFrame
print("Number of rows in combined_df:", len(combined_df))

# Check for duplicate rows in the combined DataFrame
duplicate_rows = combined_df[combined_df.duplicated()]
print("Number of duplicate rows in combined_df:", len(duplicate_rows))

# Optionally, you can display the duplicate rows
if not duplicate_rows.empty:
    print("Duplicate rows:")
    print(duplicate_rows)
else:
    print("No duplicate rows found.")
```

```python
#Impute Missing Values
import pandas as pd

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv("telecom_dataset.csv")

# Calculate the mean of the specified column
mean_value = df['total day charge'].mean()

# Fill the missing values in the specified column with the calculated mean
df['total day charge'].fillna(mean_value, inplace=True)

# Display the DataFrame after filling missing values
print("DataFrame after filling missing values:")
print(df)
```

```python
#Interpolate Missing Values
import pandas as pd

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv("telecom_dataset.csv")

# Interpolate missing values in the 'total day charge' column using linear interpolation
df['total day charge'].interpolate(method='linear', inplace=True)

# Display the DataFrame after interpolation
print("DataFrame after linear interpolation:")
print(df)
```

```python
#Handling Categorical Data
import pandas as pd

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv("telecom_dataset.csv")

# Calculate the mode of the specified column
mode_value = df['international plan'].mode()[0]

# Fill the missing values in the specified column with the mode value
df['international plan'].fillna(mode_value, inplace=True)

# Display the DataFrame after filling missing values
print("DataFrame after filling missing values:")
print(df)
```

```python
#Histograms:
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Specify the full path to the CSV file
file_path = "/telecom_dataset.csv"

# Read the data
df = pd.read_csv(file_path)

# Plot a histogram using Matplotlib for 'total day minutes'
plt.hist(df['total day minutes'], bins=10)
plt.xlabel('Total Day Minutes')
plt.ylabel('Frequency')
plt.title('Histogram of Total Day Minutes')
plt.show()

# Plot a histogram using Seaborn for 'total day calls'
sns.histplot(df['total day calls'], bins=10)
plt.xlabel('Total Day Calls')
plt.ylabel('Frequency')
plt.title('Histogram of Total Day Calls')
plt.show()
```

```python
#Scatter Plots:
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Specify the full path to the CSV file
file_path = "/telecom_dataset.csv"

# Read the data
df = pd.read_csv(file_path)

# Plot a scatter plot using Matplotlib for 'total day minutes' vs 'total day calls'
plt.scatter(df['total day minutes'], df['total day calls'])
plt.xlabel('Total Day Minutes')
plt.ylabel('Total Day Calls')
plt.title('Scatter Plot of Total Day Minutes vs Total Day Calls')
plt.show()

# Plot a scatter plot using Seaborn for 'total day minutes' vs 'total day calls'
sns.scatterplot(x='total day minutes', y='total day calls', data=df)
plt.xlabel('Total Day Minutes')
plt.ylabel('Total Day Calls')
plt.title('Scatter Plot of Total Day Minutes vs Total Day Calls')
plt.show()
```

```python
#Box Plots:
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Specify the full path to the CSV file
file_path = "/telecom_dataset.csv"

# Read the data
df = pd.read_csv(file_path)

# Plot a box plot using Matplotlib for 'total day minutes'
plt.boxplot(df['total day minutes'])
plt.xlabel('Total Day Minutes')
plt.title('Box Plot of Total Day Minutes')
plt.show()

# Plot a box plot using Seaborn for 'total day calls'
sns.boxplot(x='total day calls', data=df)
plt.xlabel('Total Day Calls')
plt.title('Box Plot of Total Day Calls')
plt.show()
```

```python
#Heatmaps:
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Specify the full path to the CSV file
file_path = "/telecom_dataset.csv"

# Read the data
df = pd.read_csv(file_path)

# Select only numerical columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot a heatmap using Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/telecom_dataset.csv"
df = pd.read_csv(file_path)

# Select only numerical columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Pairwise scatter plot
sns.pairplot(numeric_df)
plt.title('Pairwise Scatter Plot of Numerical Features')
plt.show()

# Correlation matrix heatmap
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
